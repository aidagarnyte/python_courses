```python
import pandas as pd
```

Data on problems logged through the "Tvarkau Vilnių" app.

```python
df = pd.read_csv('tvarkaumiesta_list_2022_11.csv', parse_dates=['REG_DATE'])
df.head()
```

Clean up the data

```python
unique = df.drop_duplicates('ID')
unique.head()
```

Let's find out which city registered the most complaints.

```python
df2 = unique.groupby(['city_label'])['city_label'].count()
df2
```

```python
import matplotlib.pyplot as plt

ax = unique['city_label'].value_counts().plot(kind='barh',
                                    figsize=(14,8),
                                    title="Number of complaints for each city")

for container in ax.containers:
    ax.bar_label(container)
```

Find out how many complaints were solved, and how many days on average the complaint takes to solve.

```python
non_completed = unique['COMPLETE_DATE'].isna().sum()
percentage_of_completed = (1 - (non_completed/len(unique)))*100
print(f"{percentage_of_completed.round(2)}%")
```

```python
print(f"{int(df['SOLVED_TIME'].mean().round())} days")
```

Find the most common problem type.

```python
df3 = unique.groupby(['PROBLEM_TYPE'])['PROBLEM_TYPE'].count()
df3
```

```python
ax = unique['PROBLEM_TYPE'].value_counts().plot(kind='barh',
                                    figsize=(14,8),
                                    title="Number of complaints for each type")

for container in ax.containers:
    ax.bar_label(container)
```

What day of the week municipality received the most complaints.

```python
sorter = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

dates_day_count = unique['REG_DATE'].dt.day_name().value_counts().reindex(sorter)
ax = dates_day_count.plot.line()
```
